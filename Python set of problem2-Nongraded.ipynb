Problem 2: Prime Numbers


Identify and create a list of prime numbers from the original list.

```python
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(math.sqrt(n)) + 1, 2):
        if n % i == 0:
            return False
    return True


prime_numbers = list(itertools.filterfalse(lambda x: not is_prime(x), originallist))
print(f"First 10 prime numbers: {prime_numbers[:10]}")
```

```python
# Problem 2: Prime Numbers
import math

def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(math.sqrt(n)) + 1, 2):
        if n % i == 0:
            return False
    return True

@profile
def problem_2_prime_numbers(originallist):
    prime_numbers = list(filter(is_prime, originallist))
    print(f"First 10 prime numbers: {prime_numbers[:10]}")
    return prime_numbers

problem_2_prime_numbers(originallist)
```


Problem 3: Fibonacci Sequence


Find the first 100 Fibonacci numbers that are present in the original list.

```python
def fibonacci_seq():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b


fibonacci_set = set(itertools.islice(fibonacci_seq(), 100))  # First 100 Fibonacci numbers
fib_in_list = list(itertools.filterfalse(lambda x: x not in fibonacci_set, originallist))
print(f"First 10 Fibonacci numbers in the list: {fib_in_list[:10]}")

```

```python
# Problem 3: Fibonacci Sequence
def fibonacci_seq():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

@profile
def problem_3_fibonacci_in_list(originallist):
    fibonacci_set = set(itertools.islice(fibonacci_seq(), 100))  # Get first 100 Fibonacci numbers
    fib_in_res = list(filter(lambda x: x in fibonacci_set, originallist))
    print(f"First 10 Fibonacci numbers in list: {fib_in_res[:10]}")
    return fib_in_res

problem_3_fibonacci_in_list(originallist)
```

```python
# Problem 6: Common Elements
import random

@profile
def problem_6_common_elements(originallist):
    res2 = random.sample(range(1, 5000), len(originallist))  # Second list of the same size
    common_elements = list(set(originallist).intersection(res2))
    print(f"Number of common elements: {len(common_elements)}")
    return common_elements

problem_6_common_elements(originallist)
```

```python
from functools import reduce

# Problem 15: Reduce to the Sum of Squares
def problem_15_sum_of_squares(originallist):
    # Use reduce to calculate the sum of the squares
    sum_of_squares = reduce(lambda acc, x: acc + x ** 2, originallist, 0)
    
    # Print the result
    print(f"Sum of squares: {sum_of_squares}")
    return sum_of_squares

# Example original list for testing
originallist = random.sample(range(1, 5000), 1000)

# Call the function
problem_15_sum_of_squares(originallist)
```

Problem 16: Expand

```python
# Since compression removed only consecutive duplicates, 
# we can't recreate the exact original without additional data.
# However, if we saved consecutive counts, we could expand.

compressed_with_counts = [(key, len(list(group))) for key, group in itertools.groupby(originallist)]
expanded_list = [key for key, count in compressed_with_counts for _ in range(count)]
print(f"Expanded list (first 10 elements): {expanded_list[:10]}")
```

```python
import random

# Problem 16: Expand the Compressed List (custom logic for expansion)
def problem_16_expand(compressed_list):
    # For demonstration, let's expand each element by a random count (1 to 3 times)
    expanded_list = []
    for item in compressed_list:
        # Expand the item by repeating it randomly 1 to 3 times
        expanded_list.extend([item] * random.randint(1, 3))
    
    # Print the first 10 elements of the expanded list for demonstration
    print(f"First 10 expanded elements: {expanded_list[:10]}")
    return expanded_list

# Example compressed list for testing (using Problem 15 logic)
compressed_list = [key for key, group in itertools.groupby(originallist)]

# Call the function
problem_16_expand(compressed_list)
```
